```python

drug_df['Nicotine_Score'] = drug_df['Nicotine'].str.extract('CL(\\d)').astype(float)
```

```python
# Staying indoor numeric
indoor_map = {
    'None': 0,
    '1-14 days': 1,
    '15-30 days': 2,
    '31-60 days': 3,
    '61-90 days': 4,
    '91-120 days': 5,
    'More than 120 days': 6
}
mental_health_df['Days_Indoors_Score'] = mental_health_df['Days_Indoors'].map(indoor_map)

# Stres score
mental_health_df['Stress_Score'] = mental_health_df['Growing_Stress'].map({'Yes': 1, 'No': 0})

# Nicotine
drug_df['Nicotine_Score'] = drug_df['Nicotine'].str.extract('CL(\\d)').astype(float)
```

```python
# Mental health avg
mh_grouped = mental_health_df.groupby(['Country', 'Gender'])[['Days_Indoors_Score', 'Stress_Score']].mean().reset_index()

# Drug data nicotine score
drug_grouped = drug_df.groupby(['Country', 'Gender'])[['Nicotine_Score']].mean().reset_index()

# Obesity data just last year
obesity_latest = obesity_df[obesity_df['IsLatestYear'] == True]
obesity_grouped = obesity_latest.groupby(['Location', 'Dim1'])['FactValueNumeric'].mean().reset_index()
obesity_grouped.columns = ['Country', 'Gender', 'Obesity_Rate']

#merge the tables
merged_df = pd.merge(mh_grouped, drug_grouped, on=['Country', 'Gender'], how='inner')
merged_df = pd.merge(merged_df, obesity_grouped, on=['Country', 'Gender'], how='inner')

merged_df.head()
```

```python
print("Mental health countries:", mental_health_df['Country'].unique())
print("Drug countries:", drug_df['Country'].unique())
print("Obesity countries:", obesity_df['Location'].unique())

```

```python
print("Obesity genders:", obesity_df['Dim1'].unique())
print("Mental health genders:", mental_health_df['Gender'].unique())
print("Drug genders:", drug_df['Gender'].unique())
```

```python
# gender names standardised
gender_map = {'M': 'Male', 'F': 'Female'}
drug_df['Gender'] = drug_df['Gender'].map(gender_map)

# match country names
country_map = {
    'USA': 'United States',
    'UK': 'United Kingdom',
    'Republic of Ireland': 'Ireland'
}
drug_df['Country'] = drug_df['Country'].replace(country_map)

# rename location as country in obesity data
obesity_df = obesity_df.rename(columns={'Location': 'Country', 'Dim1': 'Gender'})
```

```python
# merge again and group
mh_grouped = mental_health_df.groupby(['Country', 'Gender'])[['Days_Indoors_Score', 'Stress_Score']].mean().reset_index()
drug_grouped = drug_df.groupby(['Country', 'Gender'])[['Nicotine_Score']].mean().reset_index()
obesity_latest = obesity_df[obesity_df['IsLatestYear'] == True]
obesity_grouped = obesity_latest.groupby(['Country', 'Gender'])['FactValueNumeric'].mean().reset_index()
obesity_grouped = obesity_grouped.rename(columns={'FactValueNumeric': 'Obesity_Rate'})

merged_df = pd.merge(mh_grouped, drug_grouped, on=['Country', 'Gender'], how='inner')
merged_df = pd.merge(merged_df, obesity_grouped, on=['Country', 'Gender'], how='inner')

merged_df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(merged_df, hue='Gender', vars=['Days_Indoors_Score', 'Stress_Score', 'Nicotine_Score', 'Obesity_Rate'])
plt.suptitle("Lifestyle Factors vs Obesity by Gender", y=1.02)
plt.show()
```

```python
from scipy.stats import ttest_ind

# seperate female and male
female = merged_df[merged_df['Gender'] == 'Female']['Obesity_Rate']
male = merged_df[merged_df['Gender'] == 'Male']['Obesity_Rate']

# t-test
t_stat, p_val = ttest_ind(female, male, equal_var=False)

print("T-statistics:", round(t_stat, 3))
print("P-value:", round(p_val, 5))

if p_val < 0.05:
    print("Result: Meaningful difference between genders")
else:
    print("Result: Not that meaningful difference between genders")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter American Samoa data
samoa_df = obesity_df[obesity_df['Country'] == 'American Samoa']

# Optional: Keep only Male and Female (exclude 'Both sexes' if exists)
samoa_df = samoa_df[samoa_df['Gender'].isin(['Male', 'Female'])]

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=samoa_df, x='Gender', y='FactValueNumeric', palette='Set2')
plt.title("Obesity Rate in American Samoa by Gender")
plt.ylabel("Obesity Rate (%)")
plt.xlabel("Gender")
plt.ylim(0, samoa_df['FactValueNumeric'].max() + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

# Filter the dataset for American Samoa
samoa_df = obesity_df[obesity_df['Country'] == 'American Samoa']

#  Separate the groups by gender
female_samoa = samoa_df[samoa_df['Gender'] == 'Female']['FactValueNumeric']
male_samoa = samoa_df[samoa_df['Gender'] == 'Male']['FactValueNumeric']

#  Check the number of observations
print("♀ Number of female samples:", len(female_samoa))
print("♂ Number of male samples:", len(male_samoa))

# Perform an independent t-test (Welch's test)
t_stat, p_val = ttest_ind(female_samoa, male_samoa, equal_var=False)

#  Round results for readability
t_stat_rounded = round(t_stat, 3)
p_val_rounded = round(p_val, 5)

# Print the results
print("\nT-statistic:", t_stat_rounded)
print("P-value:", p_val_rounded)

if p_val < 0.05:
    print(" RESULT: There is a statistically significant difference in obesity rates between males and females in American Samoa.")
else:
    print("RESULT: No statistically significant difference found between genders in American Samoa.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
drug_df = pd.read_csv("/mnt/data/Drug_Consumption.csv")

# Normalize country names
country_map = {
    'USA': 'United States',
    'United States': 'United States',
    'UK': 'United Kingdom',
    'United Kingdom': 'United Kingdom',
}
drug_df['Country'] = drug_df['Country'].map(country_map).fillna(drug_df['Country'])

# Filter for specific countries
selected_countries = ['Australia', 'Canada', 'United Kingdom', 'United States']
filtered_df_nscore = drug_df[drug_df['Country'].isin(selected_countries)]

# Group by country and gender, calculate average Nscore
nscore_grouped = (
    filtered_df_nscore.groupby(['Country', 'Gender'])['Nscore']
    .mean()
    .reset_index()
)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=nscore_grouped, x='Country', y='Nscore', hue='Gender')
plt.title("Average Nscore (Neuroticism) by Country and Gender")
plt.ylabel("Average Nscore")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()

# Save plot
plt.savefig("/mnt/data/nscore_by_country_gender.png")
plt.show()
```
